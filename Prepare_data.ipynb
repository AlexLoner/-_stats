```python
import pandas as pd
import re
```

```python
df1 = pd.read_csv('data_nature', sep='@')
df2 = pd.read_csv('JETP', sep='@')
df2["Journal"] = 'JETP'
df3 = pd.read_csv('JETP_LETTERS', sep='@')
df3["Journal"] = 'JETP_LETTERS'
df4 = pd.read_csv('PHYSICAL_REVIEW_LETTERS', sep='@')
df4["Journal"] = 'PHYSICAL_REVIEW_LETTERS'
```

```python
df1.info()
```

```python
df2.info()
```

```python
df3.info()
```

```python
df4.info()
```

```python
data = pd.DataFrame(columns=['Year', 'Authors', 'Title', 'Qoutes', 'Abstract', 'Link'])
for fr in [df1, df2, df3, df4]:
    data = pd.concat((data, fr[fr.Authors != 'Incognito']), axis=0, ignore_index=True)
data = data.drop(columns='Unnamed: 0')
```

```python
data.info()
```

```python
data = data.dropna()
```

```python
data.info()
```

```python
data.Title = data.Title.str.lower()
data.Abstract = data.Abstract.str.lower()
data.Abstract = data.Abstract.str.replace('[^a-z0-9]+',' ')
data.Year = data.Year.apply(int)
data.Qoutes = data.Qoutes.apply(int)
```

```python
data = data.sort_values("Qoutes", ascending=False)
```

```python
data.to_csv('data', sep='@', index=False)
```
